```python
def remove_cols(data):
    rows = data.shape[0]
    for column in data[data.columns[4:]]:
#       print(column)
        n_zero = rows - (np.count_nonzero(data[column]))
#       print(n_zero)
        if n_zero >= (0.7 * rows):
            data.drop(columns = column, inplace = True)
            
    return data
```
